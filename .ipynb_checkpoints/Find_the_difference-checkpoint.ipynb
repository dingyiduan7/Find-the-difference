```python
from PIL import Image, ImageChops
import matplotlib.pyplot as plt
import easygui as eg
```

```python
# Open the two images to be compared
image1 = Image.open(eg.fileopenbox("Select the first image"))
image2 = Image.open(eg.fileopenbox("Select the second image"))
```

```python
# Create a figure and axis object
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10,10))

# Display the first image on the left axis
axes[0].imshow(image1)
axes[0].set_title('original image')
# Display the second image on the right axis
axes[1].imshow(image2)
axes[1].set_title('different image')


# Hide the tick marks and labels
axes[0].axis('off')
axes[1].axis('off')

# Show the figure
plt.show()
```

```python
# Get the format of the images
image1_format = image1.format
print('image1 original format is:',image1_format)
image2_format = image2.format
print('image2 original format is:',image2_format)
```

```python
# check if the image is jpeg or png, if png, convert to jpeg

# for image1
if image1_format == 'JPEG':
    print("image1 is already JPEG format")
else:
    image1 = image1.convert('RGB')
    image1.save('image1.jpg')
    
# for image2
if image2_format == 'JPEG':
    print("image2 is already JPEG format")
else:
    image2 = image2.convert('RGB')
    image2.save('image2.jpg')
```

```python
# Re-open the two images to be compared
image1 = Image.open(eg.fileopenbox("Select the first image"))
image2 = Image.open(eg.fileopenbox("Select the second image"))
```

```python
# Get the format of the images
image1_format = image1.format
print('image1 original format is:',image1_format)
image2_format = image2.format
print('image2 original format is:',image2_format)
```

```python
# Get the sizes of the images
width1, height1 = image1.size
width2, height2 = image2.size

print("image1 size:", width1, height1)
print("image2 size:", width2, height2)
```

```python
# Check if the images are the same size
if width1 != width2 or height1 != height2:
    print("The images are different sizes.")
    
else:
    print("The images are the same sizes.")
    # Create a transparent image to store the differences
    diff_image = Image.new('RGB', (width1, height1), (0, 0, 0))
    # Loop through each pixel in the images
    for x in range(width1):
        for y in range(height1):
            # Get the pixel values for each image
            pixel1 = image1.getpixel((x, y))
            pixel2 = image2.getpixel((x, y))

            # Compare the pixel values
            if pixel1 != pixel2:
                # If the pixels are different, set the pixel in the difference image to red
                diff_image.putpixel((x, y), (255, 0, 0))
```

```python
diff_image
```

Make the diff_image background transparent, then overlay on top the different image

```python
# put the diff_image on top of the second image
blended_image = Image.blend(diff_image, image2, 0.5)
blended_image
```

```python
# Create a figure and axis object
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (10,10))

# Display the first image on the left axis
axes[0].imshow(image1)
axes[0].set_title('original image')
# Display the second image on the right axis
axes[1].imshow(image2)
axes[1].set_title('different image')

# Display the blended image on the right axis
axes[2].imshow(blended_image)
axes[2].set_title('the difference')

# Hide the tick marks and labels
axes[0].axis('off')
axes[1].axis('off')
axes[2].axis('off')

# Show the figure
plt.show()
```
